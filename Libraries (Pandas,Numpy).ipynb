#Groupby function
* Instead of calculating average salaries seperately, we can use Groupby function by pandas

```python
#Groupby agrregate
```

```python
df.groupby('rank')['salary'].mean()
```

```python
#min, max, mean, median
```

```python
df.groupby('rank')['salary'].agg({'min','max','mean','median'})
```
